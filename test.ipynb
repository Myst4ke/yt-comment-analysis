```python
import re
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

sid = SentimentIntensityAnalyzer()

# Define a function to preprocess text based on language
def preprocess_text(text, lang):
    if isinstance(text, str):  # Check if the value is a string
        text = text.lower()
        text = re.sub(r'http\S+', '', text)
        text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
        tokens = word_tokenize(text)
        if lang == 'US':
            stop_words = set(stopwords.words('english'))
        elif lang == 'FR':
            stop_words = set(stopwords.words('french'))
        else:
            return ''  # Skip preprocessing for non-English and non-French comments
        tokens = [word for word in tokens if word not in stop_words]
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(word) for word in tokens]
        preprocessed_text = ' '.join(tokens)
        return preprocessed_text
    else:
        return ''

# Apply language detection and preprocessing to comments
def preprocess_comments(comment, lang):
    return preprocess_text(comment, lang)

df['comment'] = df['comment'].astype(str)

df['preprocessed_comment'] = df.apply(lambda row: preprocess_comments(row['comment'], row['region']), axis=1)
df
```

```python

# Define function to retrieve comments for a video
def get_comments(video_id):
    comments = df[df['videoID'] == video_id]['preprocessed_comment'].tolist()
    return comments

# Define function to calculate like, dislike, and neutral counts based on sentiment
def calculate_sentiment_counts(comments):
    like_count = 0
    dislike_count = 0
    neutral_count = 0
    for comment in comments:
        sentiment_score = sid.polarity_scores(comment)
        if sentiment_score['compound'] > 0.05:
            like_count += 1
        elif sentiment_score['compound'] < -0.05:
            dislike_count += 1
        else:
            neutral_count += 1
    return like_count, dislike_count, neutral_count

# Create a new DataFrame to store video IDs, positive comments, negative comments, and neutral comments
video_sentiment_df = pd.DataFrame(columns=['Video ID', 'Positive Comments', 'Negative Comments', 'Neutral Comments'])

# Process each video
for video_id in df['videoID'].unique():
    comments = df[df['videoID'] == video_id]['preprocessed_comment'].tolist()
    like_count, dislike_count, neutral_count = calculate_sentiment_counts(comments)
    video_sentiment_df = video_sentiment_df.append({'Video ID': video_id, 
                                                    'Positive Comments': like_count, 
                                                    'Negative Comments': dislike_count,
                                                    'Neutral Comments': neutral_count}, 
                                                   ignore_index=True)

# Display the new DataFrame
video_sentiment_df
```

```python
# Merge video_sentiment_df with df_video based on 'Video ID' column
merged_df = pd.merge(video_sentiment_df, df_videos, how='left', left_on='Video ID', right_on='id')

# Drop the redundant 'id' column (if needed)
merged_df.drop(columns=['id'], inplace=True)

# Drop rows with duplicate Video IDs
merged_df = merged_df.drop_duplicates(subset='Video ID', keep='first')

# Reset index after dropping rows
merged_df = merged_df.reset_index(drop=True)


merged_df
```

```python
import matplotlib.pyplot as plt

# Filter the DataFrame for French videos
fr_df = merged_df[merged_df['region'] == 'FR']

# Filter the DataFrame for US videos
us_df = merged_df[merged_df['region'] == 'US']

def plot_sentiment_counts(df, title, sort_by='Total Comments'):
    # Calculate total comments
    df['Total Comments'] = df['Positive Comments'] + df['Negative Comments'] + df['Neutral Comments']

    # Sort the DataFrame by the total number of comments
    df_sorted = df.sort_values(by=sort_by, ascending=False)
    
    positive_comments = df_sorted['Positive Comments']
    negative_comments = df_sorted['Negative Comments']
    neutral_comments = df_sorted['Neutral Comments']
    num_videos = len(df_sorted)

    # Create a bar plot
    plt.figure(figsize=(50, 30))
    plt.bar(range(num_videos), positive_comments, color='lightgreen', label='Positive Comments')
    plt.bar(range(num_videos), neutral_comments, color='lightblue', label='Neutral Comments', bottom=positive_comments)
    plt.bar(range(num_videos), negative_comments, color='lightcoral', label='Negative Comments', bottom=positive_comments + neutral_comments)
    plt.ylabel('Comment Counts')
    plt.title(title)
    plt.legend(['Positive Comments', 'Negative Comments', 'Neutral Comments'], loc='upper right')
    plt.xticks([])
    plt.tight_layout()
    plt.show()

# Plot sentiment counts for French videos, sorted by Total Comments
plot_sentiment_counts(fr_df, 'Sentiment Counts for French Videos (Sorted by Total Comments)')

# Plot sentiment counts for US videos, sorted by Total Comments
plot_sentiment_counts(us_df, 'Sentiment Counts for US Videos (Sorted by Total Comments)')
```

```python
def plot_sentiment_ratios(df, title, sort_by='Positive %'):
    # Calculate total comments
    df['Total Comments'] = df['Positive Comments'] + df['Negative Comments'] + df['Neutral Comments']
    
    # Calculate percentage of positive, negative, and neutral comments
    df['Positive %'] = (df['Positive Comments'] / df['Total Comments']) * 100
    df['Negative %'] = (df['Negative Comments'] / df['Total Comments']) * 100
    df['Neutral %'] = (df['Neutral Comments'] / df['Total Comments']) * 100

    # Sort the DataFrame by the total number of comments
    df_sorted = df.sort_values(by=sort_by, ascending=False)
    
    positive_percentage = df_sorted['Positive %']
    negative_percentage = df_sorted['Negative %']
    neutral_percentage = df_sorted['Neutral %']
    num_videos = len(df_sorted)

    # Create a bar plot
    plt.figure(figsize=(50, 30))
    plt.bar(range(num_videos), positive_percentage, color='lightgreen', label='Positive Comments')
    plt.bar(range(num_videos), neutral_percentage, color='lightblue', label='Neutral Comments', bottom=positive_percentage)
    plt.bar(range(num_videos), negative_percentage, color='lightcoral', label='Negative Comments', bottom=positive_percentage + neutral_percentage)
    plt.ylabel('Comment Percentage')
    plt.title(title)
    plt.legend(['Positive Comments', 'Neutral Comments', 'Negative Comments'], loc='upper right')
    plt.xticks([])
    plt.tight_layout()
    plt.show()

# Plot sentiment ratios for French videos, sorted by Total Comments
plot_sentiment_ratios(fr_df, 'Sentiment Ratios for French Videos (Sorted by Total Comments)')

# Plot sentiment ratios for US videos, sorted by Total Comments
plot_sentiment_ratios(us_df, 'Sentiment Ratios for US Videos (Sorted by Total Comments)')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Define English stop words
english_stopwords = set(stopwords.words('english'))

# Define a function to preprocess tags
def preprocess_tags(tags):
    if isinstance(tags, str):
        # Convert to lowercase
        tags = tags.lower()
        # Remove punctuation
        tags = re.sub(r'[^\w\s]', '', tags)
        # Tokenize the tags
        tags_tokens = word_tokenize(tags)
        # Remove stop words
        tags_tokens = [word for word in tags_tokens if word not in english_stopwords]
        # Join the tokens back into a single string
        cleaned_tags = ' '.join(tags_tokens)
        return cleaned_tags
    else:
        return ''  # Return an empty string for non-string inputs

# Apply preprocessing to the tags column
us_df['cleaned_tags'] = us_df['tags'].apply(preprocess_tags)

# Concatenate all cleaned tags into a single string
us_cleaned_text = ' '.join(us_df['cleaned_tags'])

# Generate the word cloud for US videos
us_wordcloud = WordCloud(width=800, height=400, background_color='white').generate(us_cleaned_text)

# Plot the word cloud for US videos
plt.figure(figsize=(10, 6))
plt.imshow(us_wordcloud, interpolation='bilinear')
plt.title('Word Cloud of Most Common Words in Tags for US Videos')
plt.axis('off')  # Hide axis
plt.show()
```
