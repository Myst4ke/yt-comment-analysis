```python
import re
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

sid = SentimentIntensityAnalyzer()

# Define a function to preprocess text based on language
def preprocess_text(text, lang):
    if isinstance(text, str):  # Check if the value is a string
        text = text.lower()
        text = re.sub(r'http\S+', '', text)
        text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
        tokens = word_tokenize(text)
        if lang == 'US':
            stop_words = set(stopwords.words('english'))
        elif lang == 'FR':
            stop_words = set(stopwords.words('french'))
        else:
            return ''  # Skip preprocessing for non-English and non-French comments
        tokens = [word for word in tokens if word not in stop_words]
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(word) for word in tokens]
        preprocessed_text = ' '.join(tokens)
        return preprocessed_text
    else:
        return ''

# Apply language detection and preprocessing to comments
def preprocess_comments(comment, lang):
    return preprocess_text(comment, lang)

df['preprocessed_comment'] = df.apply(lambda row: preprocess_comments(row['comment'], row['region']), axis=1)

# Define function to retrieve comments for a video
def get_comments(video_id):
    comments = df[df['videoID'] == video_id]['preprocessed_comment'].tolist()
    return comments

# Define function to calculate like and dislike counts based on sentiment
def calculate_sentiment_counts(comments):
    like_count = 0
    dislike_count = 0
    for comment in comments:
        sentiment_score = sid.polarity_scores(comment)
        if sentiment_score['compound'] > 0.05:
            like_count += 1
        elif sentiment_score['compound'] < -0.05:
            dislike_count += 1
    return like_count, dislike_count

# Create a new DataFrame to store video IDs, positive comments, and negative comments
video_sentiment_df = pd.DataFrame(columns=['Video ID', 'Positive Comments', 'Negative Comments'])

# Process each video
for video_id in df['videoID'].unique():
    comments = df[df['videoID'] == video_id]['preprocessed_comment'].tolist()
    like_count, dislike_count = calculate_sentiment_counts(comments)
    video_sentiment_df = video_sentiment_df.append({'Video ID': video_id, 
                                                    'Positive Comments': like_count, 
                                                    'Negative Comments': dislike_count}, 
                                                   ignore_index=True)

# Display the new DataFrame
video_sentiment_df
```

```python
# Merge video_sentiment_df with df_video based on 'Video ID' column
merged_df = pd.merge(video_sentiment_df, df_videos, how='left', left_on='Video ID', right_on='id')

# Drop the redundant 'id' column (if needed)
merged_df.drop(columns=['id'], inplace=True)

# Drop rows with duplicate Video IDs
merged_df = merged_df.drop_duplicates(subset='Video ID', keep='first')

# Reset index after dropping rows
merged_df = merged_df.reset_index(drop=True)


merged_df
```

```python
import matplotlib.pyplot as plt

# Filter the DataFrame for French videos
fr_df = merged_df[merged_df['region'] == 'FR']

# Filter the DataFrame for US videos
us_df = merged_df[merged_df['region'] == 'US']

def plot_sentiment_counts(df, title, sort_by='Total Comments'):
    # Calculate total comments
    df['Total Comments'] = df['Positive Comments'] + df['Negative Comments']
    
    # Sort the DataFrame by the total number of comments
    df_sorted = df.sort_values(by=sort_by, ascending=False)
    
    positive_comments = df_sorted['Positive Comments']
    negative_comments = df_sorted['Negative Comments']
    num_videos = len(df_sorted)

    # Create a bar plot
    plt.figure(figsize=(50, 30))
    plt.bar(range(num_videos), positive_comments, color='lightgreen', label='Positive Comments')
    plt.bar(range(num_videos), negative_comments, color='lightcoral', label='Negative Comments', bottom=positive_comments)
    plt.ylabel('Comment Counts')
    plt.title(title)
    plt.legend(['Positive Comments', 'Negative Comments'], loc='upper right')
    plt.xticks([])
    plt.tight_layout()
    plt.show()

# Plot sentiment counts for French videos, sorted by Total Comments
plot_sentiment_counts(fr_df, 'Sentiment Counts for French Videos (Sorted by Total Comments)', sort_by='Total Comments')

# Plot sentiment counts for US videos, sorted by Total Comments
plot_sentiment_counts(us_df, 'Sentiment Counts for US Videos (Sorted by Total Comments)', sort_by='Total Comments')
```

```python

# Define the video metric to plot (e.g., 'ViewCount', 'likeCount', 'commentCount')
video_metric = 'ViewCount'

# Define a function to create a scatter plot
def scatter_plot_sentiment_vs_metric(df, metric):
    plt.figure(figsize=(10, 6))
    plt.scatter(df[metric], df['Positive Comments'], color='lightgreen', label='Positive Comments')
    plt.scatter(df[metric], df['Negative Comments'], color='lightcoral', label='Negative Comments')
    plt.xlabel(metric)
    plt.ylabel('Comment Counts')
    plt.title(f'Sentiment vs. {metric}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Create scatter plot for French videos
scatter_plot_sentiment_vs_metric(fr_df, video_metric)

# Create scatter plot for US videos
scatter_plot_sentiment_vs_metric(us_df, video_metric)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Define English stop words
english_stopwords = set(stopwords.words('english'))

# Define a function to preprocess tags
def preprocess_tags(tags):
    if isinstance(tags, str):
        # Convert to lowercase
        tags = tags.lower()
        # Remove punctuation
        tags = re.sub(r'[^\w\s]', '', tags)
        # Tokenize the tags
        tags_tokens = word_tokenize(tags)
        # Remove stop words
        tags_tokens = [word for word in tags_tokens if word not in english_stopwords]
        # Join the tokens back into a single string
        cleaned_tags = ' '.join(tags_tokens)
        return cleaned_tags
    else:
        return ''  # Return an empty string for non-string inputs

# Apply preprocessing to the tags column
us_df['cleaned_tags'] = us_df['tags'].apply(preprocess_tags)

# Concatenate all cleaned tags into a single string
us_cleaned_text = ' '.join(us_df['cleaned_tags'])

# Generate the word cloud for US videos
us_wordcloud = WordCloud(width=800, height=400, background_color='white').generate(us_cleaned_text)

# Plot the word cloud for US videos
plt.figure(figsize=(10, 6))
plt.imshow(us_wordcloud, interpolation='bilinear')
plt.title('Word Cloud of Most Common Words in Tags for US Videos')
plt.axis('off')  # Hide axis
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Define French stop words
french_stopwords = set(stopwords.words('french'))

# Define a function to preprocess tags
def preprocess_tags(tags):
    if isinstance(tags, str):
        # Convert to lowercase
        tags = tags.lower()
        # Remove punctuation
        tags = re.sub(r'[^\w\s]', '', tags)
        # Tokenize the tags
        tags_tokens = word_tokenize(tags, language='french')
        # Remove stop words
        tags_tokens = [word for word in tags_tokens if word not in french_stopwords]
        # Join the tokens back into a single string
        cleaned_tags = ' '.join(tags_tokens)
        return cleaned_tags
    else:
        return ''  # Return an empty string for non-string inputs

# Apply preprocessing to the tags column
fr_df['cleaned_tags'] = fr_df['tags'].apply(preprocess_tags)

# Concatenate all cleaned tags into a single string
fr_cleaned_text = ' '.join(fr_df['cleaned_tags'])

# Generate the word cloud for French videos
fr_wordcloud = WordCloud(width=800, height=400, background_color='white').generate(fr_cleaned_text)

# Plot the word cloud for French videos
plt.figure(figsize=(10, 6))
plt.imshow(fr_wordcloud, interpolation='bilinear')
plt.title('Word Cloud of Most Common Words in Tags for French Videos')
plt.axis('off')  # Hide axis
plt.show()
```
